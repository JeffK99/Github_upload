```python
#1.plot a histogram of a set of observations

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
num_bins = 5

plt.hist(x, bins=num_bins, width=5)
plt.show()
```

```python
#2. Calculate statistical information of a set of observations using numpy and stats

observations = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]

#Find the mean of observations
total=sum(observations)
print(str('mean of observations ='),(total/26))

#Find the Median
x=sorted(observations)
print(x)
print(str('median of observations ='), (np.median(x)))

#Find the Mode
Count = stats.mode(x)
print(str('mode of observations ='), Count)

#Find the percentile
print(str('percentile of observations ='), (str('[')), np.percentile(observations, 25), np.percentile(observations, 50), np.percentile(observations, 75),(str(']')))

#Find the statistical summary of observations
print(str('statistical summary of observations ='), stats.describe(observations))
```

```python
#4. Create a Binomial Distribution
# creating a Binomial Distribution
n = 10               #number of trials
p = 0.5              #probability of occurrence of each trial
k = np.arange(0,20)  #deine a list of value

print(k)

binomial = stats.binom.pmf(k, n, p)  #obtain the probability mass function for a certain value of k, n and p

#print(binomial)

plt.plot(binomial);
```

```python
#5. Create a scatter plot
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii
plt.scatter(x,y,colors, area)
plt.show()
```
